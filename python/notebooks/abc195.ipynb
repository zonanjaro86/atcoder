```python
import random
```

```python
def get_test_data():
    A = random.randint(1, 1000)
    B = random.randint(1, 1000)
    W = random.randint(1, 1000)
    if A > B:
        A, B = B, A
    return A, B, W
```

```python
def solve(A, B, W):
    n_min = 1001
    n_max = 0
    for n in range(0, 1001):
        # n: みかんの個数
        if A * n <= W*1000 <= B * n:
            if n < n_min:
                n_min = n
            if n > n_max:
                n_max = n
    if n_max == 0 or n_min == 1001:
        print('UNSATISFIABLE')
    else:
        print(n_min, n_max)
```

```python
solve(A, B, W)
```

```python
for i in range(100):
    A, B, W = get_test_data()
    print(A, B, W)
    solve(A, B, W)
```
